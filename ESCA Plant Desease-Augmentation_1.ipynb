```python
import os
import cv2
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
import matplotlib.pyplot as plt
from pathlib import Path

# Define the transformations and their corresponding functions
transformation_array = [
    "horizontalFlip",
    "verticalFlip",
    "rotation",
    "widthShift",
    "heightShift",
    "shearRange",
    "zoom",
    "blur",
    "brightness",
    "contrast",
    "saturation",
    "hue",
    "gamma"
]

enable_show = False  # Set to True if you want to visualize the augmented images

# Define the paths
healthy_path = 'C:/pythonn/plant_desease/healthy/'
rust_path = 'C:/pythonn/plant_desease/rust/'

augmented_path = 'C:/pythonn/plant_desease/augmented_new/'

classes = ['healthy', 'rust']

for class_tag in classes:
    input_path = os.path.join(healthy_path if class_tag == 'healthy' else rust_path)
    output_path = os.path.join(augmented_path, class_tag)

    # Create the output directory if it doesn't exist
    os.makedirs(output_path, exist_ok=True)

    for filename in os.listdir(input_path):
        if filename.endswith(".jpg"):
            # Copy the original image to the new dataset
            original_file_path = os.path.join(input_path, filename)
            original_newname_file_path = os.path.join(output_path, Path(filename).stem + "_original.jpg")
            os.system(f"cp {original_file_path} {original_newname_file_path}")

            # Initialize the ImageDataGenerator with the chosen augmentation
            for transformation in transformation_array:
                if transformation == "horizontalFlip":
                    datagen = ImageDataGenerator(horizontal_flip=True)
                elif transformation == "verticalFlip":
                    datagen = ImageDataGenerator(vertical_flip=True)
                elif transformation == "rotation":
                    datagen = ImageDataGenerator(rotation_range=40, fill_mode='nearest')
                elif transformation == "widthShift":
                    datagen = ImageDataGenerator(width_shift_range=0.2, fill_mode='nearest')
                elif transformation == "heightShift":
                    datagen = ImageDataGenerator(height_shift_range=0.2, fill_mode='nearest')
                elif transformation == "shearRange":
                    datagen = ImageDataGenerator(shear_range=0.2)
                elif transformation == "zoom":
                    datagen = ImageDataGenerator(zoom_range=[0.5, 1.0])
                elif transformation == "blur":
                    datagen = ImageDataGenerator()
                elif transformation == "brightness":
                    datagen = ImageDataGenerator(brightness_range=[1.1, 1.5])
                elif transformation == "contrast":
                    datagen = ImageDataGenerator()
                elif transformation == "saturation":
                    datagen = ImageDataGenerator()
                elif transformation == "hue":
                    datagen = ImageDataGenerator()
                elif transformation == "gamma":
                    datagen = ImageDataGenerator()

                # Load a sample image
                img = load_img(original_file_path)
                # Convert the input image to an array
                data = img_to_array(img)
                # Reshape the input image to expand dimensions to one sample
                samples = np.expand_dims(data, 0)

                # Plot the original image
                print("Original image:")
                print(filename)
                if enable_show:
                    plt.imshow(img)
                    plt.show()
                    print("\n\n")

                # Generate and save augmented images
                print("Apply " + transformation + ".")
                # Prepare iterator
                it = datagen.flow(samples, batch_size=1, save_to_dir=output_path,
                                  save_prefix=Path(filename).stem + "_" + transformation, save_format='jpg')
                batch = it.next()
                # Plot the transformed image
                image = batch[0].astype('uint8')
                if enable_show:
                    print("Transformed image:")
                    plt.imshow(image)
                    plt.show()
                print("\n\n")

print("Done!\n\n")
```
